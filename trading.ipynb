```python
columns=["Árfolyam", "ÁrfolyamMA30", "ÁrfolyamMA90"]
```

```python
url_otp_supra_dollar_alap = "https://www.bamosz.hu/alapoldal?isin=HU0000718481"
```

```python
def convert_to_float(item):
  return float(item[0] + "." + item[1:])
```

```python
def get_data_bamosz(url):
  df = pd.read_html(url_otp_supra_dollar_alap, attrs={"class" : "dataTable2"}, decimal=",")[1]
  df.columns = df.values[0]
  df = df[1:].copy().iloc[::-1]
  df["Árfolyam"] = df["Árfolyam"].apply(lambda x: convert_to_float(x))
  df["ÁrfolyamMA30"] = df["Árfolyam"].rolling(30).mean()
  df["ÁrfolyamMA90"] = df["Árfolyam"].rolling(90).mean()
  return df.set_index("Dátum", drop=True)
```

```python
otp_supra_dollar_alap = get_data_bamosz(url_otp_supra_dollar_alap)
```

```python
sd = "2021.01.01"
ed = "2022.12.31"
otp_supra_dollar_alap[columns][(otp_supra_dollar_alap.index >= sd) & (otp_supra_dollar_alap.index <= ed)].plot(figsize=(20, 10))
```
